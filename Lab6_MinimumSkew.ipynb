# Problem1: Minimum Skew Problem 

Find a Position in a Genome Minimizing the Skew 

```python
def skew(sequence):
    c = 0
    g = 0
    minskew = 0
    skewlist = []
    index = 0
    for i in sequence:
        index += 1
        if i == 'C':
            c += 1
        if i == 'G':
            g += 1
        skew = g-c
        if skew < minskew:
            skewlist = [index]
            minskew = skew
        if skew == minskew and index not in skewlist:
            skewlist.append(index)
    print(skewlist)
```

```python
skew("")
```

# Problem 2: Implement PatternToNumbe

```python
def PatternToNumber(pattern):
    seq = {'A':0, 'C':1, 'G':2, 'T':3}
    k = len(pattern)
    num = 0
    for i in range(k):
        num += seq[pattern[i]] * pow(4, k-i-1)
    return num
```

```python
PatternToNumber("ACAAGGGTCGTAAGCAGGCCCAGGATGA")
```

# Problem 3: Implement NumberToPattern  

```python
def NumberToPattern(index, k):
    bases = ['A', 'C', 'G', 'T']
    pattern = ''
    for i in range(k):
        pattern += bases[index % 4]
        index = index // 4
    return pattern[::-1]
```

```python
NumberToPattern(5038,10)
```

# Problem 4: Finding frequent words by sorting

```python
import time
def FindingFrequentWordsBySorting(Text, k):
    FrequentPatterns = set()
    index = list()
    count = list()
    SortedIndex = list()
    t0=time.time()
    for i in range(len(Text) - k + 1):
        Pattern = Text[i:i+k]
        index.append(PatternToNumber(Pattern))
        count.append(1)
    SortedIndex = sorted(index)
    for i in range(1, len(Text)-k):
        if SortedIndex[i] == SortedIndex[i-1]:
            count[i] = count[i-1] + 1
    maxCount = max(count)
    for i in range(len(Text)-k):
        if count[i] == maxCount:
            Pattern = NumberToPattern(SortedIndex[i],k)
            FrequentPatterns.add(Pattern)
    for pattern in FrequentPatterns:
        print(pattern, end=' ')
    t1=time.time()
    t=t1-t0
 
    print('Time (seconds) :')
    print(t)
```
